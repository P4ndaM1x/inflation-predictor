## data_parser.py

```python
import subprocess
subprocess.run("python data_parser.py", shell=True)

import numpy as np
indicators_labels = np.load('./macroeconomic_data/indicators_labels.npy', allow_pickle=True).item()
subindicators_labels = np.load('./macroeconomic_data/subindicators_labels.npy', allow_pickle=True).item()
parsed_data = np.load('./macroeconomic_data/parsed_data.npy', allow_pickle=True).item()
```

```python
# format: {int(indicator_id): {int(subindicator_id): str(label), ...}, ...}
subindicators_labels
```

```python
# format: {(int(year), int(month)): {int(indicator_id): {int(subindicator_id): float(value), ...}, ...}, ...}
# cells marked with a dot in `.csv` files are represented by None values
parsed_data

for date, indicators in parsed_data.items():
    if indicators is None or indicators.keys() != indicators_labels.keys():
        print(indicators.keys(), indicators_labels.keys())
```
